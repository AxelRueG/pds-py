```python
import numpy as np
import matplotlib.pyplot as plt
from utils import Signals
```

```python
t1, y1 = Signals.senoidal(0,1,50,2,1)
_ , y2 = Signals.cuadrada(0,1,50,2)
t3, y3 = Signals.sinc(-2,2,25,2)
```

```python
ax1 = plt.subplot(221)
ax1.stem(t1,y1)
ax1.set_title('senoidal')
ax2 = plt.subplot(222)
ax2.stem(t1,y2)
ax2.set_title('cuadrada')
ax3 = plt.subplot(212)
ax3.stem(t3,y3)
ax3.set_title('sinc')

plt.tight_layout()
plt.show()
```

```python
from utils import SignalsOperations as s_op

ax1 = plt.subplot(311)
ax1.stem(t1,y1)
ax1.set_title('senial original')

ax2 = plt.subplot(323)
t1i, y1i = s_op.inversion(t1,y1) 
ax2.stem(t1i, y1i)
ax2.set_title('senial invertida')

ax3 = plt.subplot(325)
ax3.stem(t1, s_op.rectificacion_media_onda(y1))
ax3.set_title('senial rectificada media onda')

ax4 = plt.subplot(326)
ax4.stem(t1, s_op.rectificacion_onda_completa(y1))
ax4.set_title('senial rectificada onda completa')

ax5 = plt.subplot(324)
ax5.stem(t1, s_op.cuantizador(y1))
ax5.set_title('senial cuantizada')

plt.tight_layout()
```
